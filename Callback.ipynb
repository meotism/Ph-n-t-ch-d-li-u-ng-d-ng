```python
def sortArrayLesster(a):
  for x in range(0,len(a)-1):
    for y in range(x,len(a)):
      if(a[x]<a[y]):
        temp=a[x]
        a[x]=a[y]
        a[y]=temp
  return a
def sortArrayGreater(a):
  for x in range(0,len(a)-1):
    for y in range(x,len(a)):
      if(a[x]>a[y]):
        temp=a[x]
        a[x]=a[y]
        a[y]=temp
  return a
def callbackSort(a,callback):
  for x in range(0,len(a)-1):
    for y in range(x,len(a)):
      if(callback(a[x],a[y])):
        temp=a[x]
        a[x]=a[y]
        a[y]=temp
  return a
```
