```python
#Ham tinh tong
def sum_calculator(agr):
  sum = 0
  for i in agr:
    sum= sum + int(i)
  return sum
```

```python
#Ham xoa phan tu trung
def del_duplicate(a):
  for i in a:
    if a.count(i) > 1:
        a.remove(i)
  print(a)
```

```python
#1
A=['5','6','7','8']
print(A)
B=('5','6','7',8)
print(B)
#2
sum_calculator(A)
sum_calculator(B)
```
