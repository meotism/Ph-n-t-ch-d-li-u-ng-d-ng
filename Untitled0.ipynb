```python
import pandas as pd
import numpy as np
df_temp=pd.read_csv("chipotle.tsv", sep='\t', header=0)
df = pd.DataFrame (df_temp, columns = ['order_id','quantity','item_name','choice_description','item_price'])
df.head()
```

```python
df["item_price"] = df["item_price"].replace({'\$':''}, regex = True)
df.head()
```

```python
df["item_price"]=pd.to_numeric(df["item_price"])
print(df.dtypes)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.isnull().sum().sort_values(ascending=False)
for column in df.columns:
  percent=df[column].isnull().mean()
  print(f'{column}:{round(percent*100,3)}')
```

```python
df.fillna(method='bfill')
```

```python
df.isnull().sum().sort_values(ascending=False)
for column in df.columns:
  percent=df[column].isnull().mean()
  print(f'{column}:{round(percent*100,3)}')
```
