```python
#Câu 10:
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('iris')
sb.boxplot(x = "species", y = "petal_length", data = df)
plt.show()
```

Câu 10: Nhân xét: Từ biểu đồ ta thấy, setosa có 3 điểm ngoại lệ, versiclor species có 1 điểm ngoại lệ .Biểu đồ hình hộp bị lệch âm

```python

#11
import pandas as pd
import matplotlib.pyplot as plt
fig = df[df.species=='sepal_length'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='blue', label='versicolor',ax=fig)
df[df.Species=='sepal_width'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```
