```python
import pandas as pd
from scipy import stats
import statistics
import numpy as np
import requests
import json
from pandas import json_normalize 
r = requests.get('https://api.tsdh.online/api/majors/majors-3')
x = r.json()
df = pd.DataFrame(columns=["name","code","id"])
df = pd.DataFrame.from_dict(x, orient='columns')
df
df.to_csv("nhom_nganh.csv",encoding='utf-8-sig')
```

```python
r = requests.get('https://api.tsdh.online/api/group-major-interest/major-group')
x = r.json()
df = pd.DataFrame(columns=["id","name","majors"])
df = pd.DataFrame.from_dict(x, orient='columns')
df= df.iloc[:, 0:2]
dfm = json_normalize(x, 'majors')
dfm.columns=["id_majors","code","name_majors"]
result = pd.concat([df, dfm])
result.to_csv("linh_vuc.csv",encoding='utf-8-sig')
result
```
